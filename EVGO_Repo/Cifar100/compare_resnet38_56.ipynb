```python
import csv
import numpy
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
EVGO_lossepoch = np.loadtxt(open("freya_alg4_lossepoch_r38_cifar100_reduce_sum.csv", "rb"),  delimiter="," )
EVGO_acc_test = np.loadtxt(open("freya_alg4_acc_test_r38_cifar100_reduce_sum.csv", "rb"),  delimiter="," )
EVGO_loss_tra = np.loadtxt(open("freya_alg4_loss_tra_r38_cifar100_reduce_sum.csv", "rb"),  delimiter="," )

GD_lossepoch = np.loadtxt(open("freya_base_lossepoch_r38_cifar100.csv", "rb"),  delimiter="," )
GD_acc_test = np.loadtxt(open("freya_base_acc_test_r38_cifar100.csv", "rb"),  delimiter="," )
GD_loss_tra = np.loadtxt(open("freya_base_loss_tra_r38_cifar100.csv", "rb"),  delimiter="," )

Adam_lossepoch = np.loadtxt(open("thor2_Adam_lossepoch_r38_cifar100.csv", "rb"),  delimiter="," )
Adam_acc_test = np.loadtxt(open("thor2_Adam_acc_test_r38_cifar100.csv", "rb"),  delimiter="," )
Adam_loss_tra = np.loadtxt(open("thor2_Adam_loss_tra_r38_cifar100.csv", "rb"),  delimiter="," )
```

```python
import matplotlib
import matplotlib.pyplot as plt


fig = plt.figure()
x=25
plt.plot(range(x, 200), GD_acc_test[x:]*100,color='red')
plt.plot(range(x, 200), Adam_acc_test[x:]*100,  color='darkgreen')
plt.plot(range(x, 200), EVGO_acc_test[x:]*100,  color='darkblue')


fig.suptitle("Prediction Performance on Test Set")
plt.xlabel("Epoch")
plt.ylabel("Top-1 Test Accuracy")
plt.legend(['GD', 'Adam', 'EVGO'], loc='lower right')
plt.savefig("Test_acc_comparison_Resnet38_cifar100.png", dpi=600)
plt.show()
```

```python

GD= np.mean(np.array(GD_lossepoch).reshape(-1, 100), axis=1)
Adam= np.mean(np.array(Adam_lossepoch).reshape(-1, 100), axis=1)
EVGO= np.mean(np.array(EVGO_lossepoch).reshape(-1, 100), axis=1)


fig = plt.figure(figsize=(15,5))
x=150
plt.plot(range(x, GD.shape[0]), GD[x:],color='red')
plt.plot(range(x, GD.shape[0]), Adam[x:],color='darkgreen')
plt.plot(range(x, GD.shape[0]), EVGO[x:],color='darkblue')


fig.suptitle("Training Curves of AlexNet")
plt.xlabel("Iterations(k)")
plt.ylabel("Loss")
plt.legend(['GD', 'Adam', 'EVGO'], loc='upper right')
plt.savefig("Training_Curves_ResNet38_Cifar100.png", dpi=600)
plt.show()
```
