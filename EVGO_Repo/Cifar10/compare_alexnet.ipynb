```python
import csv
import numpy
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
alg1_lossepoch=[]
alg1_acc_test=[]
alg1_loss_tra=[]        


with open('alg1_acc_test.csv', 'rb') as csvDataFile:
    csvReader = csv.reader(csvDataFile,  delimiter=',', quotechar='"')
    for row in csvReader:
        alg1_acc_test.append(float(row[0]))

with open('alg1_lossepoch.csv', 'rb') as csvDataFile:
    csvReader = csv.reader(csvDataFile,  delimiter=',', quotechar='"')
    for row in csvReader:
        alg1_lossepoch.append(float(row[0]))
        
with open('alg1_loss_tra.csv', 'rb') as csvDataFile:
    csvReader = csv.reader(csvDataFile,  delimiter=',', quotechar='"')
    for row in csvReader:
        alg1_loss_tra.append(float(row[0]))
```

```python
len(alg1_loss_tra), len(alg1_lossepoch), len(alg1_acc_test)
```

```python
import matplotlib
import matplotlib.pyplot as plt

fig = plt.figure()
x=25
plt.plot(range(x, 200), alg1_acc_test[x:])
plt.plot(range(x, 200), alg2_acc_test[x:])
plt.plot(range(x, 200), alg3_acc_test[x:])

plt.plot(range(x, 200), base_acc_test[x:])



fig.suptitle("Prediction Performance on Test Set")
plt.xlabel("Epoch")
plt.ylabel("Top-1 Test Accuracy")
plt.legend(['EVGO-1', 'EVGO-2', 'EVGO-3', 'GD'], loc='lower right')
plt.savefig("Test_acc_comparison.png", dpi=600)
plt.show()
```
