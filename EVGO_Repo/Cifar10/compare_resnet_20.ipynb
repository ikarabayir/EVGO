```python
import csv
import numpy
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
EVGO_lossepoch = np.loadtxt(open("no_bn_alg4_lossepoch_r20.csv", "rb"),  delimiter="," )
EVGO_acc_test = np.loadtxt(open("no_bn_alg4_acc_test_r20.csv", "rb"),  delimiter="," )
EVGO_loss_tra = np.loadtxt(open("no_bn_alg4_acc_tra_r20.csv", "rb"),  delimiter="," )

```

```python
GD_lossepoch = np.loadtxt(open("no_bn_base_lossepoch_r20.csv", "rb"),  delimiter="," )
GD_acc_test = np.loadtxt(open("no_bn_base_acc_test_r20.csv", "rb"),  delimiter="," )
GD_loss_tra = np.loadtxt(open("no_bn_base_loss_tra_r20.csv", "rb"),  delimiter="," )

```

```python
Adam_lossepoch = np.loadtxt(open("no_bn_thor2_Adam_lossepoch_r20.csv", "rb"),  delimiter="," )
Adam_acc_test = np.loadtxt(open("no_bn_thor2_Adam_acc_test_r20.csv", "rb"),  delimiter="," )
Adam_loss_tra = np.loadtxt(open("no_bn_thor2_Adam_loss_tra_r20.csv", "rb"),  delimiter="," )
```

```python
import matplotlib
import matplotlib.pyplot as plt


fig = plt.figure()
x=25
plt.plot(range(x, 200), GD_acc_test[x:]*100,color='red')
plt.plot(range(x, 200), Adam_acc_test[x:]*100,  color='darkgreen')
plt.plot(range(x, 200), EVGO_acc_test[x:]*100,  color='darkblue')


fig.suptitle("Prediction Performance on Test Set")
plt.xlabel("Epoch")
plt.ylabel("Top-1 Test Accuracy")
plt.legend(['GD', 'Adam', 'EVGO'], loc='lower right')
plt.savefig("Test_acc_comparison_Resnet20_cifar10.png", dpi=600)
plt.show()
```

```python

GD= np.mean(np.array(GD_lossepoch).reshape(-1, 100), axis=1)
Adam= np.mean(np.array(Adam_lossepoch).reshape(-1, 100), axis=1)
EVGO= np.mean(np.array(EVGO_lossepoch).reshape(-1, 100), axis=1)


fig = plt.figure(figsize=(15,5))
x=150
plt.plot(range(x, 3126), GD[x:],color='red')
plt.plot(range(x, 3126), Adam[x:],color='darkgreen')
plt.plot(range(x, 3126), EVGO[x:],color='darkblue')


fig.suptitle("Training Curves of AlexNet")
plt.xlabel("Iterations(k)")
plt.ylabel("Loss")
plt.legend(['GD', 'Adam', 'EVGO'], loc='upper right')
plt.savefig("Training_Curves_ResNet20_Cifar10.png", dpi=600)
plt.show()
```
